```python
mode = Mtrain["Popularity"].mode()
Mtrain["Popularity"] = Mtrain["Popularity"].fillna(mode[0])
mode = Mtrain["key"].mode()
Mtrain["key"] = Mtrain["key"].fillna(mode[0])
median = Mtrain["instrumentalness"].median()
Mtrain["instrumentalness"] =  Mtrain["instrumentalness"].fillna(median)
```

```python
Mtrain["Popularity"] = pd.cut(Mtrain["Popularity"], bins=10, labels=False)
Mtest["Popularity"] = pd.cut(Mtest["Popularity"], 10, labels=False)
```

```python
from re import M
Mtrain[["key", "mode", "time_signature", "Popularity"]] = Mtrain[["key", "mode", "time_signature", "Popularity"]].astype("category")
Mtest[["key", "mode", "time_signature", "Popularity"]] = Mtest[["key", "mode", "time_signature", "Popularity"]].astype("category")
```

```python
Y = Mtrain["Class"]
X = Mtrain.drop(["Class", "Artist Name", "Track Name"], axis=1)
cat_cols = X.select_dtypes("category").columns
num_cols = X.select_dtypes(include=np.number).columns
print(cat_cols)
print(num_cols)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train[num_cols])
x_test_scaled = scaler.transform(x_test[num_cols])
```

```python
scaled_features_Mtrain = pd.DataFrame(x_train_scaled, index=x_train[num_cols].index, columns=x_train[num_cols].columns)
scaled_features_Mtest = pd.DataFrame(x_test_scaled, index=x_test[num_cols].index, columns=x_test[num_cols].columns)
Mtrain_prepro = pd.concat((scaled_features_Mtrain, x_train[cat_cols]), axis=1)
Mtest_prepro = pd.concat((scaled_features_Mtest, x_test[cat_cols]), axis=1)
```

```python
models = [
    ("log_reg", LogisticRegression(multi_class="multinomial", max_iter=1000)),
    ("naive_bay", GaussianNB()),
    ("sgd", SGDClassifier()),
    ("kn_clf", KNeighborsClassifier()),
    ("dtree", DecisionTreeClassifier()),
    ("rforest", RandomForestClassifier()),
    ("svm", SVC(kernel="linear")),
    ("svm_p", SVC(kernel="poly")),
    #("xgboost", XGBClassifier(enable_categorical=True)),
    ("adaboost", AdaBoostClassifier()),
    ("extratree", ExtraTreesClassifier()),
    ("gradboost", GradientBoostingClassifier()),
    #("gaussian_process", GaussianProcessClassifier()),
    ("mlp",  MLPClassifier()),
    ("quad_dis",  QuadraticDiscriminantAnalysis())
         ]
```

```python
%%time
df_metrics = pd.DataFrame(index=["name", "accuracy", "precision", "recall", "f1_score"])
n=0
for name, model in models:
    model.fit(Mtrain_prepro, y_train)
    y_pred = model.predict(Mtest_prepro)
    metrics = [name]
    metrics.append(accuracy_score(y_test, y_pred))
    metrics.append(precision_score(y_test, y_pred, average="weighted"))
    metrics.append(recall_score(y_test, y_pred, average="weighted"))
    metrics.append(f1_score(y_test, y_pred, average="weighted"))
    df_metrics[n] = metrics
    n += 1
```
